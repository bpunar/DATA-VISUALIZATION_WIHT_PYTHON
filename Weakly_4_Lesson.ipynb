```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

# 1. Set darkgrid style from seaborn

```python
sns.set_style("darkgrid")
```

Bar plots

```python
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]
index = ['A', 'B', 'C','D', 'E', 'F', 'G']
df = pd.DataFrame({'income': income,'expense': expense}, index=index)
df.head()
```

# 2. Make a bar plot

```python
df.plot.bar();
```

```python
df.plot(kind="bar", figsize=(10,6)); #kind=bar,pie..
df.plot(kind="barh", figsize=(10,6)); #kind=bar,pie..
df.plot(kind="line", subplots=True, figsize=(10,6)); #kind=bar,pie..
df.plot(kind="line", figsize=(10,6)); #kind=bar,pie..
df.plot(kind="pie", subplots=True, figsize=(10,6)); #kind=bar,pie..
df.plot(kind="hist", subplots=True, figsize=(10,6)); #kind=bar,pie..
```

# 3. Stack the bars

```python
df.plot(kind="bar", figsize=(8,5), stacked=True);
```

```python
fig, ax=plt.subplots(figsize=(8,5))

ax.bar(df.index, df.expense, label="expense", bottom=income)
ax.bar(df.index, df.income, label="income")

ax.legend();
```

# 3. Rotate the labels and set figsize

```python
df.plot(kind="bar", figsize=(8,5), stacked=True, rot=0);
```

```python
df.plot(kind="bar", figsize=(8,5), stacked=True)
plt.xticks(rotation=90);
```
